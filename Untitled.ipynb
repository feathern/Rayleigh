```python
import numpy
from rayleigh_diagnostics import Shell_Avgs
```

```python
sa = Shell_Avgs('00000050')
```

```python
import matplotlib.pyplot as plt
```

```python
sa.vals.shape
radius = sa.radius
nr = sa.nr
```

```python
t = sa.vals[:,0,sa.lut[501],0]
dt = sa.vals[:,0,sa.lut[507],0]
```

```python
%matplotlib inline
plt.plot(sa.radius,t)
plt.show()
```

```python

plt.plot(sa.radius,dt)
plt.show()
```
