```python
import numpy
import matplotlib.pyplot as plt
def swapread(fd,dtype='float64',count=1,swap=False):
        #Wrapper to numpy.fromfile that allows byteswapping based on Boolean swap
        if (swap):
                val = numpy.fromfile(fd,dtype=dtype,count=count).byteswap()
        else:
                val = numpy.fromfile(fd,dtype=dtype,count=count)
        if (len(val) == 1):
                val = val[0]
        return val
    
def check_endian(fd,sig,sigtype):
    # returns False if first element read from file matches sig
    # True otherwise
    chk = numpy.fromfile(fd,dtype=sigtype,count=1)
    if (chk == sig):
        return False
    else:
        return True

rfile='ell0_t_and_dtdr.dat'
qfile='ell0_t_and_dtdr_qi.dat'

bs = False

fd = open(rfile,'rb')
nr = swapread(fd,dtype='int32',count=1,swap=bs)
t_reg = swapread(fd,dtype='float64',count=nr,swap=bs)
dt_reg = swapread(fd,dtype='float64',count=nr,swap=bs)
fd.close()

fd = open(qfile,'rb')
nr = swapread(fd,dtype='int32',count=1,swap=bs)
t_q = swapread(fd,dtype='float64',count=nr,swap=bs)
dt_q = swapread(fd,dtype='float64',count=nr,swap=bs)
fd.close()

print(nr)
sizetuple=(10,3)
fig, ax = plt.subplots(figsize=sizetuple)
ax.plot(t_q)
plt.show()
```
